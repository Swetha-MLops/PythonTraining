```python
#1.Create a base class Shape with an abstract method calculate_area. Derive subclasses Square and Circle from Shape that implement the calculate_area method with their specific formulas. Write a function get_total_area that takes a list 
#of Shape objects and calculates the total area of all shapes.

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

class Square(Shape):
    def __init__(self, side_length):
        self.side_length = side_length

    def calculate_area(self):
        return self.side_length ** 2

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

def get_total_area(shapes):
    total_area = 0
    for shape in shapes:
        total_area += shape.calculate_area()
    return total_area

# Example usage
square = Square(5)
circle = Circle(3)

total_area = get_total_area([square, circle])
print("Total area of all shapes:", total_area)
```

```python
#2.Design a function greet that takes an object as an argument. If the object has a method say_hello, call that method. Otherwise, print a generic greeting.
# Use this function with objects of different classes that might or might not have a say_hello method to demonstrate polymorphism

def greet(obj):
    if hasattr(obj, 'say_hello') and callable(getattr(obj, 'say_hello')):
        obj.say_hello()
    else:
        print("Hello, nice to meet you!")

class Person:
    def say_hello(self):
        print("Hello, I'm a person.")

class Dog:
    def say_hello(self):
        print("Woof, I'm a dog.")

class Robot:
    pass  # No say_hello method

# Demonstrate polymorphism with different objects
person = Person()
dog = Dog()
robot = Robot()

greet(person)  # Output: Hello, I'm a person.
greet(dog)     # Output: Woof, I'm a dog.
greet(robot)   # Output: Hello, nice to meet you!
```

```python
#1.Create a class BankAccount with attributes like account_number (private) and balance (protected). Implement methods to deposit and withdraw funds 
#(with appropriate checks) and a public method to get the current balance.

class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.__account_number = account_number  # Private attribute
        self._balance = initial_balance         # Protected attribute

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited ${amount}. New balance: ${self._balance}")
        else:
            print("Invalid amount for deposit.")

    def withdraw(self, amount):
        if amount > 0:
            if self._balance >= amount:
                self._balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self._balance}")
            else:
                print("Insufficient funds.")
        else:
            print("Invalid amount for withdrawal.")

    def get_balance(self):
        return self._balance

# Example usage
account = BankAccount("123456789", 1000)
print("Initial balance:", account.get_balance())  # Output: Initial balance: 1000
account.deposit(500)                              # Output: Deposited $500. New balance: $1500
account.withdraw(200)                              # Output: Withdrew $200. New balance: $1300
print("Current balance:", account.get_balance())   # Output: Current balance: 1300
```

```python
#Design a class Calculator with private methods for basic operations (add, subtract, multiply, divide). Provide public 
#methods to perform these calculations on user-provided numbers.

class Calculator:
    def __init__(self):
        pass

    def __add(self, num1, num2):
        return num1 + num2

    def __subtract(self, num1, num2):
        return num1 - num2

    def __multiply(self, num1, num2):
        return num1 * num2

    def __divide(self, num1, num2):
        if num2 != 0:
            return num1 / num2
        else:
            print("Error: Division by zero.")
            return None

    def add(self, num1, num2):
        return self.__add(num1, num2)

    def subtract(self, num1, num2):
        return self.__subtract(num1, num2)

    def multiply(self, num1, num2):
        return self.__multiply(num1, num2)

    def divide(self, num1, num2):
        return self.__divide(num1, num2)

# Example usage
calculator = Calculator()
print("Addition:", calculator.add(5, 3))         # Output: Addition: 8
print("Subtraction:", calculator.subtract(10, 4)) # Output: Subtraction: 6
print("Multiplication:", calculator.multiply(7, 2))# Output: Multiplication: 14
print("Division:", calculator.divide(9, 3))       # Output: Division: 3.0
print("Division:", calculator.divide(9, 0))       # Output: Error: Division by zero. Division: None
```

Design a system to manage an inventory of products. This could involve classes like:

Challenge: Implement appropriate access modifiers (public, private, protected) for data security and encapsulation. Handle scenarios like insufficient stock during order creation and allow backorders (reservations for out-of-stock items).

```python
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.__product_id = product_id  # Private attribute
        self.name = name
        self.price = price
        self._quantity = quantity  # Protected attribute

    def get_name(self):
        return self.name

    def get_price(self):
        return self.price

    def check_stock(self, quantity):
        return self._quantity >= quantity

    def reduce_stock(self, quantity):
        if self.check_stock(quantity):
            self._quantity -= quantity
            return True
        else:
            print(f"Insufficient stock for {self.name}.")
            return False

class Order:
    def __init__(self, order_id, customer_name):
        self.__order_id = order_id  # Private attribute
        self.customer_name = customer_name
        self.items = []
        self.total_cost = 0

    def add_item(self, product, quantity):
        if product.check_stock(quantity):
            self.items.append((product, quantity))
            product.reduce_stock(quantity)
            print(f"Added {quantity} {product.get_name()} to the order.")
        else:
            print(f"Unable to add {quantity} {product.get_name()} to the order due to insufficient stock.")

    def remove_item(self, product, quantity):
        for item in self.items:
            if item[0] == product:
                self.items.remove(item)
                product._quantity += quantity  # Refund stock
                print(f"Removed {quantity} {product.get_name()} from the order.")
                break
        else:
            print(f"{product.get_name()} not found in the order.")

    def calculate_total(self):
        self.total_cost = sum([item[0].price * item[1] for item in self.items])
        return self.total_cost

# Example usage
product1 = Product(1, "Phone", 500, 10)
product2 = Product(2, "Laptop", 1000, 5)

order = Order(1, "John")
order.add_item(product1, 2)  # Output: Added 2 Phone to the order.
order.add_item(product2, 3)  # Output: Added 3 Laptop to the order.

order.remove_item(product1, 1)  # Output: Removed 1 Phone from the order.

print("Total cost:", order.calculate_total())  # Output: Total cost: 3000
```

ABSTRACTION

 Abstraction focuses on hiding the implementation details of a class and exposing only its essential functionalities. It allows you to define the "what" (behaviors) without revealing the "how" (implementation details).
Benefits:

Abstraction offers several benefits in software development:

1. **Encapsulation of Complexity**: By hiding implementation details, abstraction encapsulates the complexity of a class, making it easier to understand and use. Users interact with the class through its well-defined interface without needing to know the internal workings.

2. **Promotes Code Reusability**: Abstract classes can serve as blueprints for creating concrete implementations. By defining common behaviors in abstract classes, developers can reuse code across multiple subclasses, leading to less duplication and more maintainable code.

3. **Enhances Maintainability**: Abstraction decouples the interface from the implementation, allowing changes to be made to the underlying code without affecting how the class is used externally. This promotes better maintainability as modifications can be localized to the implementation details.

4. **Improves Modifiability**: Abstracting away implementation details makes it easier to modify and extend the functionality of a class. Developers can add new features or modify existing ones without impacting the code that relies on the class's interface.

5. **Enhances Clarity and Understandability**: Abstraction focuses on defining the essential behaviors of a class, making the interface clear and understandable for users. It hides unnecessary details, reducing cognitive overload and making it easier to comprehend and work with the code.

6. **Facilitates Testing**: Abstracting away implementation details can make it easier to write unit tests for the class. Since users interact with the class through its interface, testing becomes more straightforward, focusing on verifying the expected behavior rather than internal implementation specifics.

Overall, abstraction plays a crucial role in software development by promoting code reusability, enhancing maintainability, and improving clarity and understandability, leading to more robust and maintainable software systems.

Abstract classes and interfaces are both mechanisms in object-oriented programming used to define contracts or blueprints for classes, but they have distinct characteristics and purposes:

### Abstract Classes:

- **Definition**: An abstract class is a class that cannot be instantiated directly and may contain one or more abstract methods.
- **Purpose**: Abstract classes provide a way to define common behavior and characteristics for a group of related classes. They serve as a blueprint for concrete subclasses, which are required to provide implementations for all abstract methods defined in the abstract class.
- **Implementation**: Abstract classes can have both abstract methods (methods without implementation) and concrete methods (methods with implementation).
- **Inheritance**: Abstract classes support both method implementation and member variables. They allow for a mix of concrete and abstract methods.
- **Usage**: Abstract classes are used when there is a need to define a common interface for a group of related classes, and some methods may have default implementations.

### Interfaces:

- **Definition**: An interface is a contract that defines a set of methods that a class must implement. In many languages like Java, interfaces cannot contain method implementations.
- **Purpose**: Interfaces provide a way to define a contract for behavior without specifying how that behavior is implemented. They define a set of methods that classes must implement, promoting loose coupling and enabling polymorphic behavior.
- **Implementation**: Interfaces only contain method signatures (method declarations without implementation). They do not have member variables or method implementations.
- **Inheritance**: Interfaces support multiple inheritance, allowing a class to implement multiple interfaces. This enables a class to exhibit behavior from multiple sources.
- **Usage**: Interfaces are used when there is a need to define a contract for behavior that can be implemented by multiple unrelated classes. They facilitate loose coupling, polymorphism, and code reusability.

### Comparison:

- **Instantiation**: Abstract classes cannot be instantiated directly, while interfaces cannot be instantiated at all. They are meant to be extended or implemented by subclasses or classes, respectively.
- **Method Implementation**: Abstract classes can have both abstract and concrete methods, while interfaces only define method signatures without implementation.
- **Inheritance**: Abstract classes support single inheritance, while interfaces support multiple inheritance.
- **Usage**: Abstract classes are used for defining common behavior among related classes, while interfaces are used for defining contracts that multiple unrelated classes can implement.

In summary, abstract classes and interfaces serve different purposes in object-oriented programming. Abstract classes provide a way to define common behavior and characteristics for related classes, while interfaces define contracts for behavior that can be implemented by multiple unrelated classes. Each has its own strengths and use cases, and the choice between them depends on the specific requirements of the design.
