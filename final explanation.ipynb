```python
# Define a class to represent a Book
class Book:
    def __init__(self, title, author, genre):
        # Initialize the book with a title, author, and genre
        self.title = title
        self.author = author
        self.genre = genre

# Define a class to represent a User
class User:
    def __init__(self, name):
        # Initialize the user with a name and empty lists for reading list, history, and wish list
        self.name = name
        self.reading_list = []
        self.history = []
        self.wish_list = []

    # Method to add a book to the user's reading list
    def add_to_reading_list(self, book):
        self.reading_list.append(book)

    # Method to add a book to the user's history
    def add_to_history(self, book):
        self.history.append(book)

    # Method to add a book to the user's wish list
    def add_to_wish_list(self, book):
        self.wish_list.append(book)

    # Method to recommend books based on the page data
    def recommend_books(self, page_data):
        # Return the reading list if page_data is "page-1"
        if page_data == "page-1":
            return self.reading_list
        # Return the history if page_data is "page-2"
        elif page_data == "page-2":
            return self.history
        # Return a list of best sellers if page_data is "page-3"
        elif page_data == "page-3":
            # In a real application, you would fetch best sellers from a database or external API
            # For simplicity, I'll just return a list of best sellers here
            return [Book("Best Seller 1", "Author 1", "Fiction"), Book("Best Seller 2", "Author 2", "Non-Fiction")]
        # Return the wish list if page_data is "page-4"
        elif page_data == "page-4":
            return self.wish_list
        # Return an empty list if page_data is anything else
        else:
            return []

# Example usage
user1 = User("John")
book1 = Book("Python Programming", "Guido van Rossum", "Programming")
book2 = Book("Deep Learning", "Ian Goodfellow", "Artificial Intelligence")
book3 = Book("The Great Gatsby", "F. Scott Fitzgerald", "Classic")
book4 = Book("Clean Code", "Robert C. Martin", "Programming")

# Add books to user's lists
user1.add_to_reading_list(book1)
user1.add_to_reading_list(book2)
user1.add_to_history(book3)
user1.add_to_wish_list(book4)



    # Print out recommended books for each page
print("Page-1 Data:User started reading and continue reading")
for i, book in enumerate(user1.recommend_books("page-1"), 1):
    print(f"{i}. {book.title}")

print("\nPage-2 Data:User's history")
for i, book in enumerate(user1.recommend_books("page-2"), 1):
    print(f"{i}. {book.title}")

print("\nPage-3 Data:List of Best sellers")
for i, book in enumerate(user1.recommend_books("page-3"), 1):
    print(f"{i}. {book.title}")

print("\nPage-4 Data:Wish List")
for i, book in enumerate(user1.recommend_books("page-4"), 1):
    print(f"{i}. {book.title}")
```
