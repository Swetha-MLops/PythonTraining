```python
class Book:
    def __init__(self, title, author, genre, age):
        self.title = title
        self.author = author
        self.genre = genre
        self.age = age

class RecommendationSystem:
    def __init__(self, books):
        self.books = books

    def recommend_by_genre(self, genre):
        recommended_books = []
        for book in self.books:
            if book.genre == genre:
                recommended_books.append(book.title)
        return recommended_books

    def recommend_by_author(self, author):
        recommended_books = []
        for book in self.books:
            if book.author == author:
                recommended_books.append(book.title)
        return recommended_books

    def recommend_by_age(self, age):
        recommended_books = []
        for book in self.books:
            if book.age.lower() == age.lower():
                recommended_books.append(book.title)
        return recommended_books

books = [
    Book("Book1", "Author1", "Fiction", "Child"),
    Book("Book2", "Author2", "Non-Fiction", "Child"),
    Book("Book3", "Author3", "Thriller", "Adult"),
    Book("Book4", "Author1", "Fiction", "Child"),
    Book("Book5", "Author2", "Non-Fiction", "Child")
]

recommendation_system = RecommendationSystem(books)

genre = "Non-Fiction"
print(f"Recommended books in the {genre} genre:")
for book in recommendation_system.recommend_by_genre(genre):
    print(book)

author = "Author1"
print(f"Recommended books for {author}:")
for book in recommendation_system.recommend_by_author(author):
    print(book)

age = "Adult"
print(f"Recommended books for {age}:")
for book in recommendation_system.recommend_by_age(age):
    print(book)
```

```python
class Book:
    def __init__(self, title, author, genre, age):
        self._title = title
        self._author = author
        self._genre = genre
        self._age = age

    def get_title(self):
        return self._title

    def get_author(self):
        return self._author

    def get_genre(self):
        return self._genre

    def get_age(self):
        return self._age

class RecommendationSystem:
    def __init__(self, books):
        self._books = books

    def recommend_by_genre(self, genre):
        recommended_books = []
        for book in self._books:
            if book.get_genre() == genre:
                recommended_books.append(book.get_title())
        return recommended_books

    def recommend_by_author(self, author):
        recommended_books = []
        for book in self._books:
            if book.get_author() == author:
                recommended_books.append(book.get_title())
        return recommended_books

    def recommend_by_age(self, age):
        recommended_books = []
        for book in self._books:
            if book.get_age().lower() == age.lower():
                recommended_books.append(book.get_title())
        return recommended_books

books = [
    Book("Book1", "Author1", "Fiction", "Child"),
    Book("Book2", "Author2", "Non-Fiction", "Child"),
    Book("Book3", "Author3", "Thriller", "Adult"),
    Book("Book4", "Author1", "Fiction", "Child"),
    Book("Book5", "Author2", "Non-Fiction", "Child")
]

recommendation_system = RecommendationSystem(books)

genre = "Non-Fiction"
print(f"Recommended books in the {genre} genre:")
for book in recommendation_system.recommend_by_genre(genre):
    print(book)

author = "Author1"
print(f"Recommended books for {author}:")
for book in recommendation_system.recommend_by_author(author):
    print(book)

age = "Adult"
print(f"Recommended books for {age}:")
for book in recommendation_system.recommend_by_age(age):
    print(book)
```
