5."What is encapsulation in Python, and how is it achieved?"
A: "Encapsulation is the concept of bundling data and methods that operate on that data within
a single unit, which is the class. In Python, encapsulation is achieved through the use of
access modifiers, such as public, private, and protected, to control access to the
attributes and methods of a class. By default, all attributes and methods in Python are
public, but you can use a single underscore prefix to make them protected, and a double
underscore prefix to make them private. This helps to hide the implementation details of
a class from the outside world and ensures data integrity."

6."Can you explain the concept of abstraction in Python?"
A: "Abstraction is the concept of showing only the necessary information to the outside world while hiding the implementation details. In Python,abstraction is achieved through abstract classes and interfaces. An abstract class is a class that cannot be instantiated and is intended to be inherited by other classes. It provides a way to define an interface or a contract that must
be implemented by any subclass. Abstract classes can contain both abstract and concrete
methods. Abstract methods are methods that are declared but not implemented, and must be
implemented by any subclass. This helps to provide a way to define a common interface for
a group of related classes without exposing the implementation details."

7."What is the difference between static and instance methods in Python?"
A: "In Python, a static method is a method that belongs to a class rather than an
instance of the class. It is called on the class itself, rather than on an instance of the
class. A static method does not have access to the class or instance variables, and it
cannot modify them. On the other hand, an instance method is a method that belongs to an
instance of a class. It is called on an instance of the class, and it has access to
the instance variables. Instance methods can modify the instance variables, and they are
used to perform actions that are specific to an instance."

The key difference between a static method and a class method is that a class method
receives the class as an implicit first argument, just like an instance method receives
